```python
import matplotlib.pyplot as plt
```

```python
COLOR_COST = "salmon"
COLOR_TIME = "steelblue"
no_images = [5, 50, 100, 500, 1000]
time_taken_min_prompt = [0.97, 9.8, 20.1, 105.3, 211.7]
time_taken_min_no_prompt = [0.24, 2.3, 5.2, 26.1, 52.5]

plt.xscale("log")
plt.yscale("log")


plt.xlabel("Number of Images", color= "black", fontsize=14)
plt.ylabel("Execution Time in Minutes", color="black", fontsize=14)
plt.title("Relationship between Number of Images and Time Taken")
plt.plot(no_images, time_taken_min_prompt, "-o", color=COLOR_TIME)
plt.plot(no_images, time_taken_min_no_prompt, "-o", color="red")
plt.legend(['Prediction without Prompts', 'Prediction with Prompts'])
for n, v in zip(no_images, time_taken_min_prompt):
    if n == 5:
        plt.text(n*1.2, v*1.3, f"{str(n)} Images, \n {str(v)} Mins", ha="center", fontsize=8, color='steelblue')
    elif n == 1000:
        plt.text(n*1.2, v, f"{str(n)} Images, \n {str(v)} Mins", ha="center", fontsize=8, color='steelblue')
    else:
        plt.text(n, v*1.3, f"{str(n)} Images, \n {str(v)} Mins", ha="center", fontsize=8, color='steelblue')
for n, v in zip(no_images, time_taken_min_no_prompt):
    plt.text(n*2, v/1.3, f"{str(n)} Images, \n {str(v)} Mins", ha="center", fontsize=8, color="red")

plt.savefig("plot_1.png",
            bbox_inches ="tight",
            pad_inches = 0.2,
            transparent = True,
            facecolor ="white",
            edgecolor ='w',
            orientation ='landscape')


plt.show()
```
